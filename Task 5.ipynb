```python
#1 How many people have heart disease and how many does not?

df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['red','green'])
plt.title("Heart Disease values")
plt.xlabel("1=Heart Disease, 0=No Heart Disease")
plt.ylabel("Amount")
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Disease','No Disease'])
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.title("Male Female ratio")
plt.legend(["Male","Female"])
```

```python
#2 People of which gender have most heart disease?

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart disease frequency for genders")
plt.xlabel("1=Heart Disease, 0=No Heart Diseas")
```

```python
#3 People of which gender have which type of chest pain?

df.cp.value_counts()
```

```python
#plot bar
df.cp.value_counts().plot(kind='bar',color=['khaki','blue','pink','lightgreen'])
plt.title("Cheast pain type Vs count")
```

```python
sns.displot(x='age',data=df,bins=30)
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
#4 People with which chest pain are most pron to have heart disease?

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x='cp',data=df,hue='target')
```

```python
#5 What is the correlation between age and maximum heart rate achieved?

correlation = df['age'].corr(df['thalach'])
print("The correlation coefficient between age and maximum heart rate achieved is approximately: ")
correlation
```

```python
#6 How does the average cholesterol level differ between males and females?

avg_chol_male = df[df['sex'] == 1]['chol'].mean()
avg_chol_female = df[df['sex'] == 0]['chol'].mean()
print("The average cholesterol level for males is approximately:",avg_chol_male)
print("The average cholesterol level for females is approximately:",avg_chol_female)
```

```python
#7 How does the mean oldpeak (ST depression induced by exercise) differ across different types of chest pain?

mean_oldpeak_cp = df.groupby('cp')['oldpeak'].mean()
mean_oldpeak_cp
```
